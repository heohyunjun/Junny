```python
# if ,elif , else  
# 위에서 부터 거짓이면 아래 if문 수행

import datetime
from pytz import timezone
now = datetime.datetime.now()
now_kor = now.astimezone(timezone("Asia/Seoul"))
months = now_kor.month + 5

if 3 <= months <= 5:
  print("봄")
elif 6 <= months <= 8:
  print("여름")
elif 9 <=months <= 11:
  print("가을")
else:
  print("장겨울")


score = float(input("학점을 입력하세요"))

if score == 4.5:
  print("신 천성신")
elif score >= 4.2:
  print ("교수님의 사랑")
elif score >= 3.5:
  print("현체제의 수호신")
elif score >= 2.8:
  print("일반인")
elif score >= 2.3:
  print("일탈인")
elif score >= 1.75:
  print("오락 선구자")
else:
  print("벌레쉑기")
```

# 딕셔너리

```python
dict_a = {
    "name" : "어벤져스",
    "type" : "히어로",
    "감독" : "몰라"
}
dict_a["배우"] = ["김","천","신","허"]

print(dict_a)
print(dict_a.items())
print(dict_a["type"])
print(dict_a["배우"][1])

for key, value in dict_a.items():
  if value == "어벤져스":
    print(key)
```

```python
fruit = {
    "name" : "건조망고",
    "type" : "당절임",
    "ingredient" : ["초콜렛","망고","레몬","콜라"],
    "origin" : "한국",
    7865 : "키값이 숫자면 그냥 씀",
    True : "키가 불리안이라 따옴표 없음"
}
print("name" in fruit)
print(fruit)


for key,values in list(fruit.items()):
  if key == "name":
    del fruit[key]
  else:
    pass
print(fruit)

fruit_b = {
    "name" : "건조망고",
    "type" : "당절임",
    "ingredient" : ["초콜렛","망고","레몬","콜라"],
    "origin" : "한국",
    7865 : "키값이 숫자면 그냥 씀",
    True : "키가 불리안이라 따옴표 없음"
}
fruit_c = {}
for key,values in fruit_b.items():
  if values != "당절임":
    fruit_c[key] = values
  else:
    pass
print(fruit_c)
```

```python
# 딕셔너리를 선언합니다
character = {
    "name" : "기사",
    "level": 12,
    "items": {
        "sword" : "불꽃의 검",
        "armor" : "풀플레이트"
    },
    "skill" : ["베기","세게 배기","아주 세게 배기"]
}

print(character["items"]["sword"])

# for 반복문을 사용합니다.
```

```python
for key in character:
  if key == "items":
    for i in character[key]:
      print("{} : {}".format(i,character[key][i]))

  elif key == "skill":
    for i in character[key]:
       print("{} : {}".format(key,i))  
  else:
    print("{} : {}".format(key,character[key]))
```

```python
# 강사님

for key in character:
  if type(character[key]) is dict:
    for small_key in character[key]:
      print(small_key, ":", character[key][small_key])

  elif type(character[key]) is list:
    for item in character[key]:
      print(key, ":", item)

  else:
    print(key, ":", character[key])
```

```python
# 남준이형 

for key, value in character.items():
    if isinstance(value, dict):
        for key2 in value:
            print(key2, ':', value[key2])
    elif isinstance(value, list):
        for sk in value:
            print(key, ':', sk)
    else:
        print(key, ':', character[key])
```
