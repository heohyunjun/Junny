# 날짜시간 활용하기

```python
import datetime
from pytz import timezone
now = datetime.datetime.now()
print(now.year,"년")
print(now.month,"월")
print(now.day,"일")
print(now.hour,"시")
print(now.minute,"분")
print(now.second,"초")

now_kor = now.astimezone(timezone('Asia/Seoul'))

print(
        now_kor.year,"년",
        now_kor.month,"월",
        now_kor.day,"일",
        now_kor.hour,"시",
        now_kor.minute,"분",
        now_kor.second,"초",
     )
print('한국 시간은 {}년 {}월 {}일 {}시 {}분 {}초'.format(
                                              now_kor.year,
                                              now_kor.month,
                                              now_kor.day,
                                              now_kor.hour,
                                              now_kor.minute,
                                              now_kor.second
                                       ))
```

```python
# 오전 오후
if now.hour < 12:
  print("현재시각은 {} 오전입니다".format(now.hour))
if now.hour > 12:
  print("현재시각은 {} 오후입니다".format(now.hour))
# 사계절 
if 3 <= now.month <= 5:
  print("이번 달은 {}로 봄이다".format(now.month))
if 6 <= now.month <= 8:
  print("이번 달은 {}로 여름이다".format(now.month))
if 9 <= now.month <= 11:
  print("이번 달은 {}로 가을이다".format(now.month))
if now.month == 12 or now.month == 1:
  print("이번 달은 {}로 겨울".format(now.month))
```

# 조건문을 이용한 홀수 짝수 판별

```python
num = input("정수를 입력하시오")
last_char = num[-1]
last_num = int(last_char)
# 조건 처럼 1줄로 써야 되는데 중간에 줄을 바꾸려면? \
if last_num == 0 \
    or last_num == 2 \
    or last_num == 4 \
    or last_num == 6 \
    or last_num == 8:
    print("짝수")

if last_num == 1 \
    or last_num == 3 \
    or last_num == 5 \
    or last_num == 7 \
    or last_num == 9:
    print("홀수")

if last_char in '02468':
  print("짝수")
if last_char in '13579':
  print("홀수")
```

```python
num = int(input("정수를 입력하시오 : "))

if num % 2 == 0:
  print("짝수입니다")
if num % 2 == 1:
  print("홀수입니다")
```

# 확인 문제
- 사용자로부터 숫자 두 개를 입력받고 첫 번째 입력받은 숫자가 큰지, 
   
   두 번째 입력받은 숫자가 큰지를 구하는 프로그램을 
   
   다음 빈칸을 채워 완성해 보세요.

```python
num_a = int(input("> 1번째 숫자 : "))
num_b = int(input("> 2번째 숫자 : "))

if num_a > num_b:
  print("처음 입력한 {}가 {} 보다 큽니다".format(num_a,num_b))
if num_b > num_a:
  print("두번째로 입력한 {}가 {} 보다 큽니다".format(num_b,num_a))
if num_a == num_b:
  print("처음 입력한 {}와 두번째 입력한 {}가 같습니다".format(num_a,num_b))
```
