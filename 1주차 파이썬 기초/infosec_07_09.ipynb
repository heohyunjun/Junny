```python
def sum_all(start = 0, end = 100, step = 1):
  output = 0
  for i in range(start, end + 1,step):
    output += i
  
  return output

print("A",sum_all(0, 100, 10))
print("B",sum_all(end = 100))
print("C",sum_all(end = 100, step =2))
```

```python
# 내 잔액
my_balance = 10000

def withdarw(out_money):
  global my_balance
  if out_money > my_balance:
    return 
  
  my_balance -= out_money
  return out_money, my_balance
```

```python
result1 = withdarw(1000)
print("출금한 돈 {} , 잔액 {}".format(result1[0],result1[1]))
```

```python
result2 = withdarw(3000)
print(result2) # 결과값이 tuple로 패킹

출금액, 잔액 = withdarw(3000)
print(출금액, 잔액) # 결과값이 tuple로 패킹
```

```python
def power(item):
  return item ** 2

power(5)

list_input_a = [2, 3, 4]
# map(함수, 리스트)
list(map(power, list_input_a))

list(map(lambda x : x ** 2,list_input_a))
```

```python
# map(함수, 리스트)
# filter(함수, 리스트)
list_input_b = [1, 2, 3, 4, 5]

def power(item):
  return item * item

def under_3(item):
  return item < 3

output_a = map(power, list_input_b)
print("# map() 함수의 실행결과")
print("map(power,list_input_b : ", output_a)
print("map(power,list_input_b : ", list(output_a))
print()


output_b = filter(under_3,list_input_b)
print("# filter() 함수의 실행결과")
print("filter(under_3, output_b) : ", output_b)
print("filter(under_3, output_b) : ", list(output_b))
print()

output_b = map(under_3,list_input_b)
print("# map() 함수의 실행결과")
print("map(under_3, output_b) : ", output_b)
print("map(under_3, output_b) : ", list(output_b))
```

```python
# 다음 빈칸을 채워 매개변수로 전달된 값들을 모두 곱해서 리턴하는 가변 매개변수 함수를 만들어보세요
def mul(*values):
  mul_value = 1
  for i in values:
    mul_value *= i
  return mul_value

print(mul(5, 7, 9, 10))
```
