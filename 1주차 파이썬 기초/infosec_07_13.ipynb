# 영화 조회순 크롤링

```python
import urllib.request as req   # as 별명  //별명으로 사용
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date=20210712'
# site = urllib.request.urlopen(url)
response = req.urlopen(url)
response  # 네이버 영화 평점 퍼올 객체 생성

soup = bs(response,"html.parser") 
# 네이버 영화평점 사이트에서 내가 원하는 정보를 꺼내는 객체 __parser// 생성
soup
```

```python
title = soup.find_all("td",{'class' : 'title'})
for i in title:
  print(i.text.replace("\n",''))
print("="*100)
title = soup.find_all("div",{"class" : "tit5"})

for i in title:
  print(i.text.replace("\n",''))

print("=" * 100)

for i in title:
  print(i.find('a').text)
```

```python
title = soup.find_all("div",{"class" : "tit5"})
rank = soup.find_all("td",{"class" : "point"})
```

```python
for title, rank in zip(title,rank):
  print("영화제목 {}  평점  {} ".format(title.find('a').text, rank.text))
```

```python
for title, point in zip(title,rank):
  print(" = " * 100)
  print("title.text,strip() = ",title.text.strip())
  print("title.find('a')['href']= ",title.find('a')['href'])
  print("point = ", point.text)
```

```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
response = requests.get('http://www.weather.go.kr/weather/observation/currentweather.jsp')
# 기상청 사이트 내용 가져옴
# response 200 : 가져오기 성공
# response 404 : 없는 URL
# response 500, 505 : 서버 다운 
# request는 urllib 처럼 한번 read하고 삭제 안되고 객체 계쏙 들고있음
```

```python
soup = bs(response.text,'html.parser')

table = soup.find("table", {"class": "table_develop3"})

for tr in table.find_all('tr'):
  print("="*100)
  print(tr)
  print("="*100)
```

```python
for tr in table.findAll("tr"):
  print("="*100)
  print("tr")
  print("="*100)
  print(tr)
  print("="*100)
  print("tr.find('a') = " ,tr.find('a'))

  if tr.find('a'):
    print("tr.find('a').text = " ,tr.find('a').text)
  print("=" * 100)
```

```python
for tr in table.findAll("tr"):
  print("="*100)
  print("$" * 100)
  print(tr)
  print("$" * 100)

  if tr.find('a'):
    print("tr.find('a').text = ", tr.find('a').text)
    print('tr.findAll("td")[5]=',tr.findAll("td")[5])
    print('tr.findAll("td")[10]=',tr.findAll("td")[9])  
 
  print("="*100)
```

```python
data = []
for tr in table.findAll('tr'):
  if tr.find('a'):
    data.append([tr.find("a").text, tr.findAll("td")[5].text,tr.findAll("td")[9].text])
data
```

```python
with open("weather.csv",'w') as file:
  file.write("point,temperature,humidity")
```

```python
for point, temperature,humidity in data:
  print("point = ", point)
  print("temperature = ", temperature)
  print("humidity = ", humiditiy)
  print("="*100)
```

```python
import pandas as pd
# panda 본명대신 별명 pd 사용
pd.read_csv("weather.csv")

df = pd.read_csv("weather.csv")
# pd의 read_csv 기능을 사용하여 "weather.csv" 불러오기


df["point"] == "합천"
# point 줄이 합천이면 True  ,, 아니면 False

df[ df["point"] == "합천" ]
# df["point"] == "합천" : point칸에 저장된 값이 "합천"인 줄  , 조건 일치true , 아니면 false

df["point"].isin(["강릉","흑산도","합천"])
# point가 "강릉" ,"흑산도" ,"합천"이면 True  , 아니면 False

df [df["point"].isin(["강릉","흑산도","합천"])]
# point가 "강릉" ,"흑산도" ,"합천"이면 True  , 아니면 False로 나눠서
# True인것(강릉,흑산도,합천) 으로 df 생성   (df는 Data Frame으로 그냥 테이블(표) 같은것!)

df [  df["point"].isin( ["서울","인천","수원","대전","대구","광주","부산","울산"] ) ]
# 위와 동일
# point가 ["서울","인천","수원","대전","대구","광주","부산","울산"]이면  df생성

df = df.set_index("point")
# point 칸을 index로 설정  *@@@@@@*
# point칸에 저장된 값은 loc 함수로 쉽게 검색 가능

df.loc[["서울","대전","인천"]] 
# 위에서 isin보다 간편하게 검색 할수있음

df["humidity"].isin([70, 74 ,80])
# humidity칸에 저장된 값이 70, 74 ,80 인 줄 조건 만족하면 True, 아니면 False

df[df["humidity"].isin([70, 74 ,80])]
df = df.reset_index()  
# 기존 인덱스 point 취소 ,,,즉 3줄쯤 위에 *@@@@@*에 point로 설정한거 초기화

df = df.set_index("humidity")
# humidity칸을 index로 설정
```
