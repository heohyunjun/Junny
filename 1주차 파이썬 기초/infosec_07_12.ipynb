# 패킹, 언패킹

```python
numbers = 1, 2, 3, 4, 5 # 숫자 5개를 tuple하나에 저장 # Packing

numbers 

a, b, c, d, e = numbers # 튜플 하나의 값들이 여러 변수에 쪼개져서 저장(unpacking)

a, _ ,_ ,d, e =  numbers

a, b, *rest = numbers

a, *rest, e = numbers

```

# 가변인자 *args
# 키워드 가변인자 *kwargs

```python
def print_numbers(*args):
  print("args = ", args)
  for arg in args:
    print("=" * 100)
    print("args = " ,arg)
    print("=" * 100)
  
  print("$" * 100)

y = 10, 20, 30, 40
print_numbers(*y) # 요렇게 전송하면 packing 상태로 인덱스 0에 들어감
                  # unpacking 후 args에 대입하고싶음
                  # 그럴떄 *을 붙임 # *리스트 y의 데이터를 unpacking후 args에 대입

```

```python
def personal_info(name, age, address):
  print("name = ", name)
  print("age = ", age)
  print("address = ", address)

y = {
    "name" : "홍길동",
     "age" : "20",
     "address" : "서울"
     }

# personal_info(age=10, name='길동호', address = "서울")
personal_info(**y) # 딕셔너리 언패킹 할떄는 **사용
```

```python
def personal_info( **kwargs):
  print(" kwargs = ", kwargs)
  for key, value in kwargs.items():
    print("=" * 100)
    print("item = ",key ,value)
    print("=" * 100)
  
  print("$" * 100)

personal_info(hobby = "게임", first_name = "길동", last_name = '홍', national = '한국')
```

```python
def personal_info( **kwargs):
  print(" kwargs = ", kwargs)
  for key, value in kwargs.items():
    print("=" * 100)
    print("key = ", key )
    print("value = ", value )
    print("=" * 100)
  
  print("$" * 100)


y = {
    
    "job" : "회사원",
    "company" : "sk",
     "salary" : 4000
    
     }
personal_info(hobby = "게임", first_name = "길동", last_name = '홍', national = '한국')
personal_info(**y)
```

```python
# BeautifulSoup로 스크레핑

from bs4 import BeautifulSoup as bs

html = '''
<html>
  <body>
      <h1>스크래핑이란</h1>
      <p>웹 페이지를 분석하는 것</p>
      <p>원하는 부분을 추출하는 것</p>
  </body>
</html>
'''

soup = bs(html,'html.parser')
tag_to_h1 = soup.html.body.h1
print(tag_to_h1)
h1 = soup.html.body.h1
print(h1)

p_tag = soup.html.body.p
print(p_tag)
```
