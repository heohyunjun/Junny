```python
df3 = pd.DataFrame(
    {
        
     "수학" : [90, 80, 70],
     "영어" : [98, 89, 95],
     "음악" : [85, 95, 100],
     "체육" : [100, 90, 80]
    
    },
    index = ["서준","우현","인아"]
)
df3[["수학" ,"영어","음악"]]   # 수학,영어 ,음악 칸/열 리턴
df4 = df3[:]

df4
# df4.drop("우현")
```
