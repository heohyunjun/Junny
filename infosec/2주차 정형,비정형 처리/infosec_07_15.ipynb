# groupby


```python
df = pd.DataFrame(
    {
      "학과" : ["컴공", "보안", "보안", "컴공", "컴공"],
      "학년" : [1, 2, 1, 2, 1],
     "영어" : [90, 100, 80, 90, 70]
    },
    
)
df
```

```python
for item in df.groupby(["학과"]): #df.groupby(["학과"]) : 학과 칸에 저장된 값이 같은 줄 끼리 그룹만듬
  print("$"*100)
  print("item")
  print(item)
  print("$"*100)
```

```python
df.groupby(["학과"]).mean()
```

```python
df.groupby(["학과"]).mean()
df[["학과","영어"]].groupby("학과").mean()
df[["학과","영어"]].groupby("학과").max()
df[["학과","영어"]].groupby("학과").min()

# groupby에서 사용가능한 함수 
# count() :데이터수, sum(), max, min, mean,median,  
# var() : 분산 , # quantile : 4분위수 25%
# first() : 첫번쨰줄    # last() : 마지막 한줄
```

```python
df[["학과","영어"]].groupby("학과").get_group("컴공")
```

```python
import seaborn as sns
import numpy as np
import pandas as pd
titanic = sns.load_dataset("titanic")
```

```python
def count_null(x):
  null_val = pd.isnull(x)
  null_count = np.sum(null_val)
  return null_count
```

```python
cnull_col = titanic.apply(count_null)
print(cnull_col)
```
