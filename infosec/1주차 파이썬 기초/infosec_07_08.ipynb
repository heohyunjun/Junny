# for 반복문

```python
# 리스트를 선언합니다.
array = [273, 32, 103, 57, 52]

# 리스트에 반복문을 적용합니다.
for element in array:
  print(element)

for i in range(len(array)):
  print("{} 번째 반복 {}".format(i,array[i]))
```

```python
# reversed  : 순서를 반대로 바꿔줌
# for 반복문 반대로 반복하기

for i in range(4, 0 -1, -1):
  print("현재 반복 변수: {}".format(i))

for i in reversed(range(5)):
  print("현재 반복 변수 : {}".format(i))

print(list(reversed(range(5))))
```

```python
# -5 : 5씩감소
# -11 < i <= 100
 
for i in range(100, -11, -5):
  if i % 2 == 0:
    print("{} 마리 {}".format(i, '강아지'))
  elif i % 3 == 0:
    print("{} 마리 {}".format(i, '고양이'))
  else:
    print("{} 마리 {}".format(i, '사자'))
```

```python
# 빈칸을 채워 키와 값으로 이루어진 각 리스트를 조합해 하나의 딕셔너리를 만들어 보세요

key_list = ["name", "hp", "mp", "level"]
value_list = ["기사", 200, 30, 5]
character = {}
character_b = {}

for i,k in zip(key_list,value_list):
  character[i] = k


for i in range(len(key_list)):
  character_b[key_list[i]] = value_list[i]



print(character)
print(character_b)
```

```python
# 1부터 숫자를 하나씩 증가시키면서 더하는 경우를 생각해 보자.
# 몇을 더할 떄 1000을 넘는지 구해라. 그리고 그떄의 값도 출력해라
# 다음은 1000이 넘는 경우를 구한 예

limit = 1000
i = 1
sum_value = 0

while True:
  sum_value += i
  print(sum_value)
  if sum_value > 10000:
    print("{}를 더할 떄 {}을 넘으며 그떄의 값은 {} 입니다".format(i,limit,sum_value))
    break

  i +=1

```

```python
# 함수에 매개변수 만들기

def print_n_times(value, n):
  for i in range(n):
    print(value)
print_n_times("안녕하세요", 5)
```

```python
# 가변 매개변수

def print_n_times(n, *values):
  for i in range(n):
    for value in values:
      print(value)
    print()

print_n_times(3,"안녕하세요", "즐거운", "파이썬 프로그래밍")

```

```python
# 기본 매개변수

def print_n_times(*value, n = 2):
  for i in range(n):
    print(value)
print_n_times("안녕하세요","안녕하세요",n =  4)
```
