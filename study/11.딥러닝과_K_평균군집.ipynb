```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import cv2
import os, glob, shutil
```

```python
path = '/content/drive/MyDrive/Colab Notebooks/성신신성/jj/data/pets'
glob_dir = path +'/*.jpg'
```

```python
# glob으로 파일 하나씩 로드후, csv2로 읽고 크기 변환
images =[cv2.resize(cv2.imread(file), (224, 224)) for file in glob.glob(glob_dir)]
```

```python
# 타입변환 차원 변환
images = np.array(np.float32(images).reshape(len(images), -1)/255)
```

```python
images
```

```python
# 사용할 모델 MobileNetV2
# include_top = False
# : 네트워크 최상단에 완전연결층은ㄹ 넣을지 여부를 선택

# weights = 'imagenet'
# : None과 imagenet 선택 가능
model =tf.keras.applications.MobileNetV2(include_top = False, weights = 'imagenet',
                                         input_shape =(224, 224, 3))
```

```python
predictions = model.predict(images.reshape(-1 ,224, 224, 3))

pred_images = predictions.reshape(images.shape[0], -1)
```
