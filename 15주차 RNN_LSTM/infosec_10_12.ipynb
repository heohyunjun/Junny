# 네이버 영화 감정 분석

```python
train_path = '/content/drive/MyDrive/Colab Notebooks/AI_허현준/15주차 RNN_LSTM/datasets/ratings_train.txt'
test_path = '/content/drive/MyDrive/Colab Notebooks/AI_허현준/15주차 RNN_LSTM/datasets/ratings_test.txt'
```

```python
with open(
    train_path, 
    'r', 
    encoding = "utf8"
    ) as f:
    print("f.read()")
    print("=" * 100)
    print(f.read()) # 처음부터 끝가지 읽고 리턴
# 너무 길어서 안나옴
```

```python
with open(
    train_path, 
    'r', 
    encoding = "utf8"
    ) as f:
    for line in f.read().splitlines(): # 줄단위로 끊어서 리스트에 넣음
      print("line=",line)
      print("line.split('\t')=",line.split('\t'))
      print("=" * 100)
```

```python
with open(
    train_path, 
    'r', 
    encoding = "utf8"
    ) as f:
    result = [line.split("\t") for line in f.read().splitlines()]
    for i in result:
      print(i)
```
